# HW 3 ข้อ 2
Analysis of Products Based on High-Value Customer Behavior

**Problem Statement:**  
การวิเคราะห์ผลิตภัณฑ์ตามพฤติกรรมของลูกค้า High-Value
และพฤติกรรมการซื้อของลูกค้า High-Value ในยุโรปในช่วงคริสต์มาสและปีใหม่
(ที่กำหนดจากยอดการซื้อรวมสูงสุด) มักจะซื้อผลิตภัณฑ์ประเภทไหนร่วมกันในช่วงเทศกาลและซื้อสินค้าอะไรแตกต่างจากลูกค้าทั่วไป?  

**Objective:**  
วางแผนโปรโมชั่นและข้อเสนอที่ออกแบบมาเฉพาะสำหรับลูกค้า High-Value

## Define High-Value Customers
High-Value Customers are typically defined as those who contribute the most to the total revenue.




Calculate the total purchase amount for each customer.

```python
# Calculate total purchase amount for each customer
data_test['TotalPurchase'] = data_test['Quantity'] * data_test['UnitPrice']
customer_total_spending = data_test.groupby('CustomerID')['TotalPurchase'].sum().reset_index()
customer_total_spending
```

- Identify the top 10% of customers based on total purchase amount as High-Value Customers.

```python
# Identify High-Value Customers (top 10%)
high_value_threshold = customer_total_spending['TotalPurchase'].quantile(0.9)
high_value_customers = customer_total_spending[customer_total_spending['TotalPurchase'] >= high_value_threshold]['CustomerID']
high_value_customers
```

## Filter Data for Festive Seasons
- Focus on transactions during the festive season (e.g., December for Christmas and New Year).

```python
# Filter data for festive season (December)
festive_data = data_test[(data_test['month'] == 12) & (data_test['day'].between(1, 31))]
festive_data
```

```python
# Separate High-Value and General Customers
high_value_festive_data = festive_data[festive_data['CustomerID'].isin(high_value_customers)]
general_festive_data = festive_data[~festive_data['CustomerID'].isin(high_value_customers)]
```

##  Analyze Purchasing Behavior

- Prepare Transaction Data

```python
# Group transactions by InvoiceNo and collect StockCode as lists
high_value_transactions = high_value_festive_data.groupby('InvoiceNo')['StockCode'].apply(list)
high_value_transactions
```

```python
general_transactions = general_festive_data.groupby('InvoiceNo')['StockCode'].apply(list)
general_transactions
```

- Clean Transaction Data
- replace non-string values (e.g., `NaN`) in the transaction data with `Unknown`.

```python
def clean_transactions(transactions):
    cleaned_transactions = []
    for transaction in transactions:
        cleaned_transaction = [item if isinstance(item, str) else "Unknown" for item in transaction]
        cleaned_transactions.append(cleaned_transaction)
    return cleaned_transactions

high_value_transactions_cleaned = clean_transactions(high_value_transactions)
general_transactions_cleaned = clean_transactions(general_transactions)
```

- Generate Association Rules

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

def get_association_rules(transactions, min_support=0.02):
    # Encode transactions
    te = TransactionEncoder()
    te_ary = te.fit(transactions).transform(transactions)
    df = pd.DataFrame(te_ary, columns=te.columns_)

    # Generate frequent itemsets
    frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)

    if frequent_itemsets.empty:
        print("No frequent itemsets found. Consider lowering min_support.")
        return pd.DataFrame()

    # Generate association rules
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
    return rules.sort_values('confidence', ascending=False)

# Generate rules for High-Value and General Customers
high_value_rules = get_association_rules(high_value_transactions_cleaned, min_support=0.015)
general_rules = get_association_rules(general_transactions_cleaned, min_support=0.015)
```

```python
# Display the number of rules
print(f"Number of High-Value Customer Rules: {len(high_value_rules)}")
print(f"Number of General Customer Rules: {len(general_rules)}")
```

### Focus on High-Value Customer Rules

>Identify Top Rules

- Sort the rules by lift to prioritize the most meaningful associations.

```python
high_value_rules_sorted = high_value_rules.sort_values(by='lift', ascending=False)
high_value_rules_sorted = high_value_rules_sorted.head(10)  # Display top 10 rules
high_value_rules_sorted
```

### Refine General Customer Rules

>Filter Rules by Confidence and Lift

- To reduce the number of rules and focus on the most significant ones, filter the rules based on confidence and lift

```python
general_rules_filtered = general_rules[
    (general_rules['confidence'] > 0.7) & (general_rules['lift'] > 1.5)
]
print(f"Number of Filtered General Customer Rules: {len(general_rules_filtered)}")
```

> Identify Common Patterns

- Look for rules that appear frequently across General Customers to understand overall market trends.

```python
general_rules_sorted = general_rules_filtered.sort_values(by='support', ascending=False)
general_rules_sorted = general_rules_sorted.head(10) # Display top 10 rules
general_rules_sorted
```

```python
# Compare unique product associations between High-Value and General Customers
unique_high_value_associations = high_value_rules_sorted[
    ~high_value_rules_sorted['antecedents'].isin(general_rules_sorted['antecedents'])
]

# Print unique associations
print("Unique High-Value Customer Holiday Product Associations:")
for idx, row in unique_high_value_associations.iterrows():
    antecedents = list(row['antecedents'])
    consequents = list(row['consequents'])
    print(f"If bought {antecedents}, likely to buy {consequents}")
```

```python
import pandas as pd

# Function to filter out nan values from descriptions
def clean_descriptions(descriptions):
    return {desc for desc in descriptions if pd.notna(desc)}

# Loop through each row in the unique_high_value_associations DataFrame
for idx, row in unique_high_value_associations.iterrows():
    antecedents = list(row['antecedents'])
    consequents = list(row['consequents'])

    # Print the association
    print(f"If bought {antecedents}, likely to buy {consequents}")

    # Find and print descriptions for antecedents
    print("Descriptions for antecedents:")
    for item in antecedents:
        descriptions = finditem(item, '')
        descriptions = clean_descriptions(descriptions)  # Clean descriptions
        print(f"  StockCode {item}: {', '.join(descriptions) if descriptions else 'No description available'}")

    # Find and print descriptions for consequents
    print("Descriptions for consequents:")
    for item in consequents:
        descriptions = finditem(item, '')
        descriptions = clean_descriptions(descriptions)  # Clean descriptions
        print(f"  StockCode {item}: {', '.join(descriptions) if descriptions else 'No description available'}")

    # Alternatively, use finditems for multiple items at once
    print("Descriptions for antecedents (using finditems):")
    descriptions = finditems(*antecedents)
    descriptions = clean_descriptions(descriptions)  # Clean descriptions
    print(f"  {', '.join(descriptions) if descriptions else 'No descriptions available'}")

    print("Descriptions for consequents (using finditems):")
    descriptions = finditems(*consequents)
    descriptions = clean_descriptions(descriptions)  # Clean descriptions
    print(f"  {', '.join(descriptions) if descriptions else 'No descriptions available'}")

    print("\n" + "="*50 + "\n")  # Add a separator for better readability
```

## Segment by Region

- Analyze High-Value Customer Behavior by Continent

```python
def analyze_continent_behavior(continent):
    continent_high_value = high_value_festive_data[
        high_value_festive_data['Continent'] == continent
    ]
    return continent_high_value['Description'].value_counts().head()

# Compare holiday product preferences across continents
for continent in ['Europe', 'North America', 'Asia']:
    print(f"\n{continent} High-Value Customer Top Products:")
    print(analyze_continent_behavior(continent))
```
