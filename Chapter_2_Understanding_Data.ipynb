```python
import pandas as pd
data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])
df
```

```python
# prompt: draw scatter plot of the df, draw bold line X1 and x2 axes

import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df['X1'], df['X2'])
plt.axhline(y=0, color='black', linestyle='--')  # Draw a horizontal line at y=0
plt.axvline(x=0, color='black', linestyle='--')  # Draw a vertical line at x=0

# Customize the plot
plt.xlabel('X1', fontweight='bold')  # Bold X1 axis label
plt.ylabel('X2', fontweight='bold')  # Bold X2 axis label
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: calculate distance matrix of the df

import pandas as pd
from scipy.spatial.distance import pdist, squareform

# Assuming 'df' is your DataFrame (replace with your actual DataFrame)
# Example DataFrame (replace with your actual DataFrame)
data = [[1, 2], [-1, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])

# Calculate the distance matrix
distance_matrix = squareform(pdist(df[['X1', 'X2']]))

distance_matrix
```

```python
# prompt: calculate distance matrix of the df use L1-norm

# Calculate the distance matrix using L1-norm (Manhattan distance)
distance_matrix_l1 = squareform(pdist(df[['X1', 'X2']], 'cityblock'))

distance_matrix_l1
```

# Normalization (Z-score & Min-Max normalization)

* ถ้าสเกลมันต่างกัน --> มันจะสนใจแค่สเกลใหญ่ ๆ --> ทำ Normalization

## Z-score

```python
# prompt: calculate z-score for df2

from scipy.stats import zscore

# Assuming df2 is your DataFrame
# Calculate z-scores for all columns in df2
df2_zscore = df2.apply(zscore)

df2_zscore
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
#plt.plot([-3,1],[0,0], color='black', linewidth=2)
#plt.plot([0,0],[-2,4], color='black', linewidth=2)
plt.plot([-1.5,1],[0,0], color='black', linewidth=2)
plt.plot([0,0],[-1.5,1], color='black', linewidth=2)
plt.scatter(df2_zscore['X1'], df2_zscore['X2'],c='r')

min_val = min(df2_zscore['X1'].min(), df2_zscore['X2'].min())
max_val = max(df2_zscore['X1'].max(), df2_zscore['X2'].max())
plt.xlim(min_val, max_val)
plt.ylim(min_val, max_val)

# Customize the plot
plt.xlabel('X1', fontweight='bold')
plt.ylabel('X2', fontweight='bold')
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: do min-max normalization for the df2 for each column separately

# Assuming df2 is your DataFrame
# Min-Max normalization for each column separately
df2_minmax = df2.copy()

for column in df2_minmax.columns:
    min_val = df2_minmax[column].min()
    max_val = df2_minmax[column].max()
    df2_minmax[column + '_normalized'] = (df2_minmax[column] - min_val) / (max_val - min_val)

df2_minmax
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.plot([-0.5,1.5],[0,0], color='black', linewidth=2)
plt.plot([0,0],[-0.5,1.5], color='black', linewidth=2)
# Access the correct column name (X1_normalized)
plt.scatter(df2_minmax['X1_normalized'], df2_minmax['X2_normalized'],c='r')


# Customize the plot
plt.xlabel('X1_normalized', fontweight='bold') # Updated x-axis label
plt.ylabel('X2_normalized', fontweight='bold')
plt.title('Scatter Plot of X1_normalized vs X2_normalized') # Updated title
plt.grid(True)

# Show the plot
plt.show()
```
