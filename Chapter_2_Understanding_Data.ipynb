# Distance

```python
import pandas as pd
data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])
df
```

```python
# prompt: draw scatter plot of the df, draw bold line X1 and x2 axes

import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df['X1'], df['X2'])
plt.axhline(y=0, color='black', linestyle='--')  # Draw a horizontal line at y=0
plt.axvline(x=0, color='black', linestyle='--')  # Draw a vertical line at x=0

# Customize the plot
plt.xlabel('X1', fontweight='bold')  # Bold X1 axis label
plt.ylabel('X2', fontweight='bold')  # Bold X2 axis label
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: calculate distance matrix of the df

from scipy.spatial.distance import pdist, squareform

# Assuming 'df' is your DataFrame with relevant columns (e.g., 'X1', 'X2')
# Replace 'X1', 'X2' with the actual column names in your DataFrame.
if 'X1' in df.columns and 'X2' in df.columns:
    distance_matrix = squareform(pdist(df[['X1', 'X2']]))
    print(distance_matrix)
else:
    print("Error: Columns 'X1' and 'X2' not found in DataFrame.")
```

# Normalization (Z-score & Min-Max normalization)

* ถ้าสเกลมันต่างกัน --> มันจะสนใจแค่สเกลใหญ่ ๆ --> ทำ Normalization

## Z-score

```python
# prompt: calculate z-score for df2

from scipy.stats import zscore

# Assuming df2 is your DataFrame
# Calculate z-scores for all columns in df2
df2_zscore = df2.apply(zscore)

df2_zscore
```

```python
# prompt: plot scatter plot df2_zscore *** เดี๋ยวกลับมาแก้โค้ด

import matplotlib.pyplot as plt

# Assuming df2_zscore is already calculated as in your provided code
plt.figure(figsize=(8, 6))
plt.scatter(df2_zscore['X1'], df2_zscore['X2'])
plt.xlabel('X1 (Z-score)', fontweight='bold')
plt.ylabel('X2 (Z-score)', fontweight='bold')
plt.axhline(y=0, color='black', linestyle='--')  # Draw a horizontal line at y=0
plt.axvline(x=0, color='black', linestyle='--')  # Draw a vertical line at x=0
plt.title('Scatter Plot of Z-scores')
plt.grid(True)
plt.show()
```

```python
# prompt: do min-max normalization for the df2 for each column separately

# Assuming df2 is your DataFrame
# Min-Max normalization for each column separately
df2_minmax = df2.copy()

for column in df2_minmax.columns:
    min_val = df2_minmax[column].min()
    max_val = df2_minmax[column].max()
    df2_minmax[column + '_normalized'] = (df2_minmax[column] - min_val) / (max_val - min_val)

df2_minmax
```

```python
# prompt: plot scatter plot
# df2_minmax

import matplotlib.pyplot as plt

# Assuming df2_minmax is already calculated as in your provided code
plt.figure(figsize=(8, 6))
plt.scatter(df2_minmax['X1_normalized'], df2_minmax['X2_normalized'])
plt.xlabel('X1 (Min-Max Normalized)', fontweight='bold')
plt.ylabel('X2 (Min-Max Normalized)', fontweight='bold')
plt.axhline(y=0, color='black', linestyle='--')  # Draw a horizontal line at y=0
plt.axvline(x=0, color='black', linestyle='--')  # Draw a vertical line at x=0
plt.title('Scatter Plot of Min-Max Normalized Data')
plt.grid(True)
plt.show()
```
